```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()
ax = plt.axes()

```

```python
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, x+1, linestyle='solid')
plt.plot(x, x-1, linestyle='dashed')
```

```python
# Limiting your graphs / plots

plt.plot(x,np.sin(x))
plt.xlim(-3,3)
plt.ylim(-0.5,2)
```

```python
plt.axis('tight')
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

```python
plt.plot(x,1.3*np.sin(x))
plt.title("An unsual sine curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```
